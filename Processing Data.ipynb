```python
import pandas as pd
import numpy as np

df = pd.read_csv("Mines_Africa_Districts.csv", header=0, encoding='utf-8')
print(df.head())
```

```python
regions = pd.read_csv("country_region_africa.csv", header=0, encoding='utf-8')
print(regions)
```

```python
df
df['countryname'].value_counts().count()
```

```python
nonsums = ['countrycode', 'ADM1', 'ADM2']
nomapdf = df.drop(nonsums, axis=1)
print(nomapdf)
```

```python
mergedf = nomapdf.groupby('countryname').sum().reset_index()
print(mergedf)
```

```python
trial= pd.merge(regions,mergedf, on='countryname', how='outer')
print(trial)
```

```python
empty = ['zinc_mine', 'uranium_mine', 'strontium_mine','sulfur_mine','mica_mine', 'boronborates_mine']

mines = trial.drop(empty, axis=1)
```
